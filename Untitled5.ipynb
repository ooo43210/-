```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon

INF = 1000 # 그림판 크기
LEN = 100 # 처음

def draw_axis(ax):
    ax.axhline(0, color='black', alpha=0.3) # x-axis line
    ax.axvline(0, color='black', alpha=0.3) # y-axis line
    ax.set_xlim(-INF, INF)
    ax.set_ylim(-INF, INF)
    ax.set_aspect('equal') # 정사각형 모양으로 backboard 그리기

def draw_rectangle(ax, p1, p2, p3, p4):
    my_polygon = Polygon([p1, p2, p3, p4])
    ax.add_patch(my_polygon)

def draw_object(ax):
    p1 = np.array([0, 0])
    p2 = np.array([LEN, 0])
    p3 = np.array([LEN, LEN])
    p4 = np.array([0, LEN])

    # 변환 수행
    q1 = translate(p1, -LEN/2, -LEN/2)
    q2 = translate(p2, -LEN/2, -LEN/2)
    q3 = translate(p3, -LEN/2, -LEN/2)
    q4 = translate(p4, -LEN/2, -LEN/2)

    s = 5
    q1 = scale(q1, s, s)
    q2 = scale(q2, s, s)
    q3 = scale(q3, s, s)
    q4 = scale(q4, s, s)

    theta = np.pi/4.0
    q1 = rotate(q1, theta)
    q2 = rotate(q2, theta)
    q3 = rotate(q3, theta)
    q4 = rotate(q4, theta)

    q1 = reflect_x(q1)
    q2 = reflect_x(q2)
    q3 = reflect_x(q3)
    q4 = reflect_x(q4)

    q1 = reflect_y(q1)
    q2 = reflect_y(q2)
    q3 = reflect_y(q3)
    q4 = reflect_y(q4)

    q1 = reflect_origin(q1)
    q2 = reflect_origin(q2)
    q3 = reflect_origin(q3)
    q4 = reflect_origin(q4)

    q1 = reflect_line(q1)
    q2 = reflect_line(q2)
    q3 = reflect_line(q3)
    q4 = reflect_line(q4)

    draw_rectangle(ax, q1, q2, q3, q4)

def translate(p, dx, dy):
    # 동차좌표계 (homogeneous coordinate)으로 변환
    hp = np.array([p[0], p[1], 1.0])
    # 이동 변환 행렬 생성
    T = np.array([[1, 0, dx], [0, 1, dy], [0, 0, 1]])
    # 행렬과 벡터의 곱
    hq = np.dot(T, hp)
    # 다시 (x, y) 좌표계로 변환
    q = np.array([hq[0]/hq[2], hq[1]/hq[2]])
    return q

def scale(p, sx, sy):
    # 동차좌표계 (homogeneous coordinate)으로 변환
    hp = np.array([p[0], p[1], 1.0])
    # 크기 변환 행렬 생성
    S = np.array([[sx, 0, 0], [0, sy, 0], [0, 0, 1]])
    # 행렬과 벡터의 곱
    hq = np.dot(S, hp)
    # 다시 (x, y) 좌표계로 변환
    q = np.array([hq[0]/hq[2], hq[1]/hq[2]])
    return q

def rotate(p, theta):
    # 동차좌표계 (homogeneous coordinate)으로 변환
    hp = np.array([p[0], p[1], 1.0])
    # 회전 변환 행렬 생성
    R = np.array([[np.cos(theta), -np.sin(theta), 0], [np.sin(theta), np.cos(theta), 0], [0, 0, 1]])
    # 행렬과 벡터의 곱
    hq = np.dot(R, hp)
    # 다시 (x, y) 좌표계로 변환
    q = np.array([hq[0]/hq[2], hq[1]/hq[2]])
    return q

def reflect_x(p):
    # 동차좌표계 (homogeneous coordinate)으로 변환
    hp = np.array([p[0], p[1], 1.0])
    # x축 반전 변환 행렬 생성
    Rx = np.array([[1, 0, 0], [0, -1, 0], [0, 0, 1]])
    # 행렬과 벡터의 곱
    hq = np.dot(Rx, hp)
    # 다시 (x, y) 좌표계로 변환
    q = np.array([hq[0]/hq[2], hq[1]/hq[2]])
    return q

def reflect_y(p):
    # 동차좌표계 (homogeneous coordinate)으로 변환
    hp = np.array([p[0], p[1], 1.0])
    # y축 반전 변환 행렬 생성
    Ry = np.array([[-1, 0, 0], [0, 1, 0], [0, 0, 1]])
    # 행렬과 벡터의 곱
    hq = np.dot(Ry, hp)
    # 다시 (x, y) 좌표계로 변환
    q = np.array([hq[0]/hq[2], hq[1]/hq[2]])
    return q

def reflect_origin(p):
    # 동차좌표계 (homogeneous coordinate)으로 변환
    hp = np.array([p[0], p[1], 1.0])
    # 반전 변환 행렬 생성
    R = np.array([[-1, 0, 0], [0, -1, 0], [0, 0, 1]])
    # 행렬과 벡터의 곱
    hq = np.dot(R, hp)
    # 다시 (x, y) 좌표계로 변환
    q = np.array([hq[0]/hq[2], hq[1]/hq[2]])
    return q

def reflect_line(p):
    # 동차좌표계 (homogeneous coordinate)으로 변환
    hp = np.array([p[0], p[1], 1.0])
    # 반전 변환 행렬 생성
    R = np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])
    # 행렬과 벡터의 곱
    hq = np.dot(R, hp)
    # 다시 (x, y) 좌표계로 변환
    q = np.array([hq[0]/hq[2], hq[1]/hq[2]])
    return q




# 그림판 선언 및 크기 설정
fig, ax = plt.subplots(figsize=(7,7))
# 축 그리기
draw_axis(ax)
# 물체 그리기
draw_object(ax)
plt.show()
```
